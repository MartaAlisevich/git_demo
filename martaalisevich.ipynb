# 13. Выведите количество игр которая сыграла каждая команда в данном сезоне?

```python
data.query('season == "regular season"')\
.groupby(['away_team'],as_index=False)\
.aggregate({'home_team_outcome': "count"})\
.rename(columns={"home_team_outcome":"count_of_matches_played"})
```

# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
winner = data.query('season == "regular season"')\
.groupby(['away_team'],as_index=False)\
.aggregate({'home_team_win': "sum"})\
.sort_values("home_team_win", ascending = False)
winner.iloc[[0]]
```

```python
#наибольшее количество матчей в сезоне набрала команда San Diego Padres
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

```python
#добаляю новый столбец с переводом в шкалу по Цельсию
data['temperature_celsius_scale'] = (data['temperature'].apply(lambda x:( x - 32)/1.8)).astype(int)
data.query('season == "regular season"').groupby(['temperature_celsius_scale'], as_index = False)\
.aggregate({'total_runs': "sum"})
```
